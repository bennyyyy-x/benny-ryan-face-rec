```python
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import matplotlib.pyplot as plt

datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
img = load_img('./benny_faces/face18.jpg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

plt.imshow(x[0] / 255)
plt.axis('off')  # This turns off the axis
plt.show()
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20, 5))  # Create a single figure with 4 subplots

i = 0
for batch in datagen.flow(x, batch_size=1):
    axs[i].imshow(batch[0].astype('uint8'))
    axs[i].axis('off')  # Turn off the axis for a cleaner look
    i += 1
    if i % 4 == 0:
        break

plt.show()
```
